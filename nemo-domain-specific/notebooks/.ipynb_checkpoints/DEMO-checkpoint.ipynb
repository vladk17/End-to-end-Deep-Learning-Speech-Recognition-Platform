# 1. Select Audio file
None of the models have *seen* this audio file before.

```python
audio_file= '../../demo/4k0c0301.wav'
ipd.Audio(audio_file) # audio playback
```

# 2. Pre-Trained Acoustic Model - Greedy Decoder

```python
MODEL_YAML = '/tmp/NeMo/examples/asr/configs/quartznet15x5.yaml'
CHECKPOINT_ENCODER = '../../demo/quartznet15x5/JasperEncoder-STEP-247400.pt'
CHECKPOINT_DECODER = '../../demo/quartznet15x5/JasperDecoderForCTC-STEP-247400.pt'
```

```python
transcript, audio, features, encoded_features, probs, preds = offline_inference(MODEL_YAML, CHECKPOINT_ENCODER, CHECKPOINT_DECODER, audio_file)
transcript
```

# 3. Fine-Tuned Acoustic Model - Greedy Decoder

```python
MODEL_YAML = '/tmp/NeMo/examples/asr/configs/quartznet15x5.yaml'
CHECKPOINT_ENCODER = '../../demo/finetuned/JasperEncoder-STEP-174000.pt'
CHECKPOINT_DECODER = '../../demo/finetuned/JasperDecoderForCTC-STEP-174000.pt'
```

```python
f_transcript, f_audio, f_features, f_encoded_features, f_probs, f_preds = offline_inference(MODEL_YAML, CHECKPOINT_ENCODER, CHECKPOINT_DECODER, audio_file)
f_transcript
```

# 4. Comparing Results

Text Comparison. Marked on **green** text insert, on **red** text delete, on **blue** text replace.

### Comparison of log probability distribution over characters

```python
plt.figure(figsize=[14,8])
plt.subplot(211)
plt.imshow(probs.transpose())
plt.axis('tight')
plt.ylabel('char idx')
plt.colorbar()
plt.title('[PRETRAINED] Log probability distribution over characters in time')
plt.gca().invert_yaxis()

plt.subplot(212)
plt.imshow(f_probs.transpose())
plt.axis('tight')
plt.ylabel('char idx')
plt.xlabel('timesteps')
plt.colorbar()
plt.title('[FINETUNED] Log probability distribution over characters in time')
plt.gca().invert_yaxis()
```
